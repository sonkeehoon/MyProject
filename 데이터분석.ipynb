```python
import seaborn as sns

plt.figure(figsize=(10, 6))

w = 0.7
sns.barplot(subway_by_day, x="요일", y="승객수", width = w, palette='winter', hue="요일")

plt.title('요일별 지하철 승객(2022)')
plt.xlabel('요일')
plt.ylabel('승객 수')
plt.show()
```

- 노선별 지하철 혼잡도

```python
subway_by_line.head()
```

```python
# 1. 노선별 평균 승차·하차 승객 수 구하기
average_boarding = subway_by_line.groupby('노선명')['승차총승객수'].mean()
average_alighting = subway_by_line.groupby('노선명')['하차총승객수'].mean()

# 2. 월별 평균 승객수 구하기
subway_by_line['사용일자'] = pd.to_datetime(subway_by_line['사용일자'], format='%Y%m%d')
subway_by_line['사용월'] = subway_by_line['사용일자'].dt.to_period('M')  # 월 단위 변환
monthly_boarding = subway_by_line.groupby('사용월')['승차총승객수'].mean() # 월 별 평균 승차 승객수
monthly_boarding_scaled = monthly_boarding / 1000


# 시각화: 1x2 배치
plt.figure(figsize=(16, 8))

# 1. 노선별 평균 승차·하차 승객수(다중 막대 그래프)
plt.subplot(1, 2, 1)
width = 0.35
x = range(len(average_boarding))
plt.bar(x, average_boarding, width, label='승차', color='skyblue')
plt.bar([p + width for p in x], average_alighting, width, label='하차', color='orange')
plt.title('지하철 노선별 승/하차수 평균(2022)')
plt.xlabel('노선')
plt.ylabel('평균 승객 수')
plt.xticks([p + width / 2 for p in x], average_boarding.index, rotation = 90)
plt.grid(axis='y', alpha=0.5)
plt.legend()

# 2. 월별 평균 승차·하차 
plt.subplot(1, 2, 2)
plt.bar(monthly_boarding_scaled.index.astype(str), monthly_boarding_scaled.values, color='lightgreen')
plt.title('지하철 승/하차 월별 평균 (단위: 천 명)', fontsize=14)
plt.xlabel('월', fontsize=12)
plt.ylabel('월별 평균 승객수 (천 명)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.5)

plt.tight_layout()
plt.show()
```
